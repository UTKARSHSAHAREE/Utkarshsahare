```python
def plot_vector2d(vector2d,origin=[0,0],**option):
    return plt.arrow(origin[0],origin[1],vector2d[0],vector2d[1],
                     head_width=0.3,head_length=0.5,**option)
```

```python
u=np.array([8,8])
v=np.array([3,5])
plot_vector2d(u,color="purple")
plot_vector2d(v,color="red")
plt.axis=([0, 10, 0, 10])
plt.grid()
plt.show()
```

```python
a=np.array([4, 7, 2])
b=np.array([2, 7, 4])
```

```python
subplot3d=plt.subplot(111,projection="3d")
x_coordinates,y_coordinates,z_coordinates=zip(a,b)
subplot3d.scatter(x_coordinates,y_coordinates,z_coordinates)
subplot3d.set_zlim3d=([0, 10])
plt.show()
```

```python
def plot_vector3d(ax, vector3d, z0, **options):
  for v in vector3d:
        x,y,z=v
        ax.plot([x,x],[y,y],[z0,z], color="black", linestyle="dashed",marker="x")
  x_coordinates,y_coordinates,z_coordinates=zip(*vector3d)
  ax.scatter(x_coordinates, y_coordinates, z_coordinates, **options)

subplot3d=plt.subplot(111,projection="3d")
subplot3d.set_zlim([0, 10])
plot_vector3d(subplot3d,[a,b], 0,color=("r","b"))
plt.show()
```

```python
u_normalised=u/LA.norm(u)
proj=v.dot(u_normalised)*u_normalised

plot_vector2d(u,color="red")
plot_vector2d(v,color="blue")

plot_vector2d(proj,color="b",linestyle=":")
plt.plot(proj[0],proj[1],"ko")
plt.plot([proj[0], v[0]], [proj[1], v[1]], "b:")

plt.text(1, 2, "$proj_u v$", color="k", fontsize=18)
plt.text(1.8, 0.2, "$v$", color="b", fontsize=18)
plt.text(0.8, 3, "$u$", color="r", fontsize=18)

plt.xlim([0.0, 10])
plt.ylim([0.0, 10])
plt.gca().set_aspect("equal")
plt.grid()
```

# angle between u and v(in radians)

```python
def vector_angle(u, v):
    cos_theta = u.dot(v) / LA.norm(u) / LA.norm(v)
    return np.arccos(cos_theta.clip(-1, 1))

theta = vector_angle(u, v)
print("Angle =", theta, "radians")
print("      =", theta * 180 / np.pi, "degrees")
```
