```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=np.array([0.9,4,5])
v=np.array([3.5,1.9,6])
```

```python
x_coordinates,y_coordinates=zip(*zip(u,v))
plt.scatter(x_coordinates,y_coordinates,color=["green"])
plt.axis([0,9,0,6])
plt.grid()
plt.show()
```

```python
def plot_vector2d(vector2d,origin=[0,0],**option):
    return plt.arrow(origin[0],origin[1],vector2d[0],vector2d[1],
                     head_width=0.3,head_length=0.5,**option)
```

```python
u=np.array([8,8])
v=np.array([3,5])
plot_vector2d(u,color="purple")
plot_vector2d(v,color="red")
plt.axis=([0, 10, 0, 10])
plt.grid()
plt.show()
```
