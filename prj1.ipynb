# TO HAVE A STABLE TRAIN TEST SPLIT EVEN AFTER UPDATING THE DATASET

```python
from zlib import crc32
import numpy as np
def is_id_in_test_set(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32
```

```python
def split_data_with_id_hash(data, test_ratio, id_column):
  ids = data[id_column]
  in_test_set =ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
  return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()
train_set, test_set = split_data_with_id_hash(housing_with_id, 0.2, "index")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

   # Calculate residuals
residuals = y_test - predictions

   # Create residual plot
plt.figure(figsize=(10, 6))
sns.residplot(x=y_test, y=residuals, lowess=True, line_kws={'color': 'red'})  # lowess smoother for trend
plt.xlabel('Actual Prices')
plt.ylabel('Residuals')
plt.title('Residual Plot for NuSVR Estimator')
plt.show()
```

This graph is called a Residual Plot. It's a visual tool used to assess how well a regression model (NuSVR model)

fits the data.
For each data point, the model makes a prediction. The residual is the difference between the actual value and the predicted value.

If the residuals are randomly scattered around the horizontal line (zero), it suggests that the model is a good fit. This means the errors are unpredictable and not systematically related to the actual values.

If there are patterns in the residuals like curvature or trend, it indicates that the model might not be capturing some underlying relationship in the data

**We can see a slight curve in the residuals, suggesting that the NuSVR model might not be perfectly capturing the non-linear relationship in the data.**
